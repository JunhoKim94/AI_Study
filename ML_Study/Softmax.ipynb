```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv("./heart/Heart_Train.csv")
```

```python
s1 = set()
for i in a.ChestPain:
    s1.add(i)
s1 = list(s1)
for i in range(len(a.ChestPain)):
    for j in range(len(s1)):
        if a.ChestPain[i] == s1[j]:
            a.ChestPain[i] = j
```

```python
s1 = set()
for i in a.Thal:
    s1.add(i)
s1 = list(s1)
for i in range(len(a.Thal)):
    for j in range(len(s1)):
        if a.Thal[i] == s1[j]:
            a.Thal[i] = j
```

```python
data = np.array(a)
x_data = data[:,0:-1]
y_data = data[:,-1]
for i in range(len(y_data)):
    if y_data[i] == 'No':
        y_data[i] = 0.0
    else:
        y_data[i] = 1.0
y_data = y_data.reshape(235,1)
x_data = np.array(x_data,dtype=np.float32)
```

```python
num = 13
X = tf.placeholder("float",[None,num])
Y = tf.placeholder("float",[None,1])
```

```python
W = tf.Variable(tf.random_normal([num,1]))
b = tf.Variable(tf.random_normal([1]))
```

```python
hypothesis = tf.nn.softmax(tf.matmul(X,W)+b)
print(hypothesis)
cost = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(hypothesis),axis = 1))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(2001):
        sess.run(optimizer, feed_dict= {X:x_data,Y:y_data})
        if step %200 ==0:
            print(step, sess.run([optimizer,cost],feed_dict={X:x_data,Y:y_data}))
```

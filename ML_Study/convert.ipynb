```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv("./heart/Heart_Test.csv")
```

```python
s1 = set()
for i in a.ChestPain:
    s1.add(i)
s1 = list(s1)
for i in range(len(a.ChestPain)):
    for j in range(len(s1)):
        if a.ChestPain[i] == s1[j]:
            a.ChestPain[i] = j
s1 = set()
for i in a.Thal:
    s1.add(i)
s1 = list(s1)
for i in range(len(a.Thal)):
    for j in range(len(s1)):
        if a.Thal[i] == s1[j]:
            a.Thal[i] = j
s1 = set()
for i in a.AHD:
    s1.add(i)
s1 = list(s1)
for i in range(len(a.AHD)):
    for j in range(len(s1)):
        if a.AHD[i] == s1[j]:
            a.AHD[i] = j
```

```python
a.to_csv("./heart_test.csv")
```

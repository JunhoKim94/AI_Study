```python
import numpy as np
```

```python
class Matrix():
    def __init__(self,goal,start):
        self.goal = goal #Target Vector
        self.start = start #Start Vector
        self.stack =[] #Children을 저장할 stack
        self.visited =set()#방문한 Node를 저장할 stack
        self.stack.append(self.start) # Start Vector를 추가
        self.Open = [1] #최대 stack 길이를 찾기위한 Stack
        self.Level = [0] #Depth를 측정하기 위한 stack
    def find(self,data):
        a = data
        for i in range(len(data)):
            for j in range(len(data)):
                if a[i][j] == 0:
                    return i,j
    def Span_Child(self,d):
        data = d.copy()
        level = self.Level[-1]
        self.visited.add(data)
        del self.stack[-1],self.Level[-1]
        x,y = self.find(data)
        Near = []
        near = np.array([[x,y-1],[x,y+1],[x-1,y],[x+1,y]])
        a = (near >=0) == (near <= len(self.start)-1)
        for i in range(len(near)):
            if all(a[i]) == True:
                Near.append(near[i])
        for i in range(len(Near)):
            x1 = Near[i][0]
            y1 = Near[i][1]
            data2 = data.copy()
            data2[x][y] = data2[x1][y1]
            data2[x1][y1] = 0
            #print(data2)
            #print(self.visited)
            for s in self.visited:
                if np.all(data2 == s):
                    judge = False
                    break
                else:
                    judge = True
            if judge:
                self.stack.append(data2)
                self.Level.append(level+1) #Mother Node의 레벨 +1
                
            if len(self.stack)>=self.Open[0]:
                del self.Open[0]
                self.Open.append(len(self.stack))#스택의 최대값을 Open Stack에 계속 갱신
                
```

```python
p= np.array([[4,3,1],
             [6,5,2],
             [0,7,8]])
q = np.array([[1,2,3],
              [4,5,6],
              [7,8,0]])

i=0
z= Matrix(q,p)
while(1):
    i = i+1
    if np.size(z.stack) ==0:
        print("fail")
        break
    z.Span_Child(z.stack[-1])
    #print(z.stack)
    #print("----------cycle end-----------")
    if np.all(z.stack[-1] == q):
        print("congratulation")
        break
print(z.stack[-1],i,z.Open,z.Level[-1])
```

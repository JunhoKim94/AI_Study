```python
import heapq
```

```python
class Node:
    def __init__(self,state,depth,path,cost):
        self.state = state
        self.depth = depth
        self.path = path
        self.cost = cost
```

```python
def move_right(node):
    child = node[:]
    blank = child.index(0)
    if blank in [ 2 , 5 , 8]:
        return False
    else:
        child[blank] = child[blank+1]
        child[blank+1] = 0
        return child
def move_left(node):
    child = node[:]
    blank = child.index(0)
    if blank in [ 0 , 3 , 6]:
        return False
    else:
        child[blank] = child[blank-1]
        child[blank-1] = 0
        return child
def move_up(node):
    child = node[:]
    blank = child.index(0)
    if blank in [ 0 , 1 , 2]:
        return False
    else:
        child[blank] = child[blank-3]
        child[blank-3] = 0
        return child
def move_down(node):
    child = node[:]
    blank = child.index(0)
    if blank in [ 6 , 7 , 8]:
        return False
    else:
        child[blank] = child[blank+3]
        child[blank+3] = 0
        return child
```

```python
def Span_Child(current,cost_func,stack,i):
    
    if type(cost_func) == int:
        if (move_right(current.state)!= False):
            a = Node(move_right(current.state),current.depth+1,current.path+"R->",current.cost+1+cost_func)
            stack.append(a)
        if (move_left(current.state) != False):
            a = Node(move_left(current.state),current.depth+1,current.path+"L->",current.cost+1+cost_func)
            stack.append(a)
        if (move_up(current.state)!= False):
            a = Node(move_up(current.state),current.depth+1,current.path+"U->",current.cost+1+cost_func)
            stack.append(a)
        if (move_down(current.state) != False):
            a = Node(move_down(current.state),current.depth+1,current.path+"D->",current.cost+1+cost_func)
            stack.append(a)
            
    elif type(cost_func) == type(Span_Child):
        if (move_right(current.state)!= False):
            temp = move_right(current.state)
            a = Node(temp,current.depth+1,current.path+"R->",current.cost+1+cost_func(temp)-cost_func(current.state))
            heapq.heappush(stack,(a.cost,i,a))
            i = i+1
        if (move_left(current.state) != False):
            temp = move_left(current.state)
            a = Node(temp,current.depth+1,current.path+"L->",current.cost+1+cost_func(temp)-cost_func(current.state))
            heapq.heappush(stack,(a.cost,i,a))
            i = i+1
        if (move_up(current.state)!= False):
            temp = move_up(current.state)
            a = Node(temp,current.depth+1,current.path+"U->",current.cost+1+cost_func(temp)-cost_func(current.state))
            heapq.heappush(stack,(a.cost,i,a))
            i = i+1
        if (move_down(current.state) != False):
            temp = move_down(current.state)
            a = Node(temp,current.depth+1,current.path+"D->",current.cost+1+cost_func(temp)-cost_func(current.state))
            heapq.heappush(stack,(a.cost,i,a))
            i = i+1
        return i
```

```python
def DFS(start,goal):
    depth_limit = 50
    visited = set()
    stack = []
    Max = 1
    node = Node(start,0,"Init->",0)
    stack.append(node)
    while(True):
        #print(stack)
        current = stack.pop()#후입선출
        v =''
        for i in range(9):
            v = str(current.state[i])+ v #string으로 변환하여 set형태에 저장 -> search속도향상

        if current.state == goal:
            current.path = current.path+"Finish"
            return current,visited,Max
        
        if current.depth < depth_limit and v not in visited:
            
            Span_Child(current,0,stack,0)
            visited.add(v)
            
        if Max < len(stack):#stack의 최대값 갱신
            Max = len(stack)
```

```python
def BFS(start,goal):
    depth_limit = 100
    visited = set()
    stack = []
    Max = 1
    node = Node(start,0,"Init->",0)
    stack.append(node)
    while(True):
        #print(stack)
        current = stack.pop(0)#선입선출
        v =''
        for i in range(9):
            v = str(current.state[i])+ v #string으로 변환하여 set형태에 저장 --> search 속도 향상

        if current.state == goal:
            current.path = current.path+"Finish"
            return current,visited,Max
        
        if current.depth < depth_limit and v not in visited:
            
            Span_Child(current,0,stack,0)
            visited.add(v)
            
        if Max < len(stack):#stack의 최대값 갱신
            Max = len(stack)
```

```python
def A_star(start,goal,cost_func):
    visited = {}#dictionary 형태로 Cost 값과 같이 저장
    stack = []
    Max = 1
    node = Node(start,0,"Init->",H1_tile(start))
    #stack.put((node.cost,node))
    j = 0
    heapq.heappush(stack,(node.cost,j,node))#j를 추가해서 정렬할때 node간 오류가 없도록 조정
    while(True): 
        current = heapq.heappop(stack)[2]
        v =''
        for i in range(9):
            v = str(current.state[i])+ v#string으로 변환하여 dictionary형태에 저장 -->search 속도 + cost비교
            
        if current.state == goal:
            current.path = current.path+"Finish"
            return current,visited,Max
        
        if v not in visited:
            
            j = Span_Child(current,cost_func,stack,j)
            visited[v] = current.cost
            
        elif v in visited and visited[v]>current.cost:
            
            j = Span_Child(current,cost_func,stack,j)
            visited[v] = current.cost
        if Max < len(stack):#stack의 최대값 갱신
            Max = len(stack)
```
